```python
BASE_PATH = "/content/drive/MyDrive/bigdata_final_project"

DATASET_PATH = f"{BASE_PATH}/dataset"
RAW_PATH = f"{BASE_PATH}/raw"
LOG_PATH = f"{BASE_PATH}/logs"
WAREHOUSE_PATH = f"{BASE_PATH}/warehouse"

os.makedirs(RAW_PATH, exist_ok=True)
os.makedirs(LOG_PATH, exist_ok=True)
os.makedirs(WAREHOUSE_PATH, exist_ok=True)
```

```python
def write_log(message, filename):

    with open(f"{LOG_PATH}/{filename}", "a") as f:
        f.write(message + "\n")
```

```python
def extract_etl_source1():

    start_time = time.time()

    source_file = f"{DATASET_PATH}/whoop_fitness_dataset_100k.csv"

    raw_destination = f"{RAW_PATH}/whoop_fitness_raw.csv"

    # copy raw data
    shutil.copy(source_file, raw_destination)

    # load data
    df = pd.read_csv(source_file)

    execution_time = time.time() - start_time

    file_size = os.path.getsize(source_file) / (1024 * 1024)

    log_message = f"""
    ===== EXTRACT SOURCE 1 =====
    Source: WHOOP Dataset CSV
    Rows: {df.shape[0]}
    Columns: {df.shape[1]}
    File Size: {file_size:.2f} MB
    Execution Time: {execution_time:.2f} seconds
    Timestamp: {datetime.now()}
    """

    write_log(log_message, "extract_log.txt")

    print("SOURCE 1 SUCCESSFULLY EXTRACTED")

    return df
```

Query 1

```python
query1 = """
SELECT
    AVG(recovery_score) AS avg_recovery
FROM fact_fitness
"""

print("QUERY 1 - Average Recovery Score")
display(pd.read_sql(query1, conn))
```

Query 2 - Average Sleep Hours by Gender

```python
query2 = """
SELECT
    u.gender,
    AVG(f.sleep_hours) AS avg_sleep_hours
FROM fact_fitness f
JOIN dim_user u
    ON f.user_id = u.user_id
GROUP BY u.gender
"""

print("QUERY 2 - Average Sleep Hours by Gender")
display(pd.read_sql(query2, conn))
```

Query 3 - Recovery Score by Fitness Level

```python
query3 = """
SELECT
    u.fitness_level,
    AVG(f.recovery_score) AS avg_recovery
FROM fact_fitness f
JOIN dim_user u
    ON f.user_id = u.user_id
GROUP BY u.fitness_level
ORDER BY avg_recovery DESC
"""

print("QUERY 3 - Recovery by Fitness Level")
display(pd.read_sql(query3, conn))
```

Query 4 - Average Calories Burned by Activity Type

```python
query4 = """
SELECT
    activity_type,
    AVG(calories_burned) AS avg_calories
FROM fact_fitness
GROUP BY activity_type
ORDER BY avg_calories DESC
"""

print("QUERY 4 - Calories Burned by Activity")
display(pd.read_sql(query4, conn))
```

Query 5 - Weekend vs Weekday Recovery

```python
query5 = """
SELECT
    d.is_weekend,
    AVG(f.recovery_score) AS avg_recovery
FROM fact_fitness f
JOIN dim_date d
    ON f.date = d.date
GROUP BY d.is_weekend
"""

print("QUERY 5 - Weekend vs Weekday Recovery")
display(pd.read_sql(query5, conn))
```

Query 6 - Average HRV by Primary Sport

```python
query6 = """
SELECT
    u.primary_sport,
    AVG(f.hrv) AS avg_hrv
FROM fact_fitness f
JOIN dim_user u
    ON f.user_id = u.user_id
GROUP BY u.primary_sport
ORDER BY avg_hrv DESC
"""

print("QUERY 6 - Average HRV by Sport")
display(pd.read_sql(query6, conn))
```

Query 7 - Highest Average Day Strain by Activity

```python
query7 = """
SELECT
    activity_type,
    AVG(day_strain) AS avg_day_strain
FROM fact_fitness
GROUP BY activity_type
ORDER BY avg_day_strain DESC
"""

print("QUERY 7 - Highest Strain Activities")
display(pd.read_sql(query7, conn))
```

Query 8 - Daily Sleep Trend

```python
query8 = """
SELECT
    date,
    AVG(sleep_hours) AS avg_sleep
FROM fact_fitness
GROUP BY date
ORDER BY date
"""

print("QUERY 8 - Sleep Trend by Date")
display(pd.read_sql(query8, conn))
```

Metadataa

Transform

```python
elt_df = pd.read_sql(
    "SELECT * FROM raw_whoop_data",
    conn_elt
)
```

```python
elt_df['workout_time_of_day'] = (
    elt_df['workout_time_of_day']
    .fillna('No Workout')
)
```

```python
elt_df['date'] = pd.to_datetime(
    elt_df['date']
)
```

Join di warehouse

```python
weather_raw = pd.read_sql(
    "SELECT * FROM raw_weather_data",
    conn_elt
)

elt_df['temperature'] = weather_raw.loc[0,'temperature']
elt_df['humidity'] = weather_raw.loc[0,'humidity']
```

Agregasi analitik

```python
agg_recovery = (
    elt_df
    .groupby('fitness_level')
    .agg({
        'recovery_score':'mean',
        'sleep_hours':'mean',
        'hrv':'mean'
    })
    .reset_index()
)
```

```python
agg_recovery.to_sql(
    "agg_recovery_by_fitness",
    conn_elt,
    if_exists="replace",
    index=False
)
```

```python
agg_activity = (
    elt_df
    .groupby('activity_type')
    .agg({
        'recovery_score':'mean',
        'day_strain':'mean',
        'calories_burned':'mean'
    })
    .reset_index()
)
```

```python
agg_activity.to_sql(
    "agg_activity_metrics",
    conn_elt,
    if_exists="replace",
    index=False
)
```

Feature Engineering Berbasis Warehouse

```python
elt_df['recovery_gap'] = (
    elt_df['recovery_score']
    - elt_df['sleep_performance']
)
```

```python
elt_df['hrv_delta'] = (
    elt_df['hrv']
    - elt_df['hrv_baseline']
)
```

```python
elt_df['rhr_delta'] = (
    elt_df['resting_heart_rate']
    - elt_df['rhr_baseline']
)
```

```python
elt_df['training_load_ratio'] = (
    elt_df['activity_strain']
    /
    (elt_df['day_strain'] + 1)
)
```

```python
elt_df['sleep_debt'] = (
    8 - elt_df['sleep_hours']
)
```

```python
elt_df.to_sql(
    "elt_final_dataset",
    conn_elt,
    if_exists="replace",
    index=False
)
```
